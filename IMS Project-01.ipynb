```python
import json
import random
```

```python
with open("inventory.json","r") as fd:
    txt = fd.read()
inventory = json.loads(txt)
with open("sales.json","r") as fd:
    txt = fd.read()
sales = json.loads(txt)
# sales = txt

def calculate_Bill(pri, quan, profit,discount):
    disc_amount = pri*quan*(discount/100)
    to_pay = pri*quan - disc_amount
    return [f"Price : {pri}, Qty: {quan}\nTotal Cost: Rs. {pri*quan}\nMoney saved due to Discount({discount}%) : Rs. {disc_amount}\nTotal amount to Pay: {to_pay}", to_pay, profit*quan]



type(inventory) # dictionary
inventory
```

```python
name = input("Enter customer Name(who is purchasing): ")
prodID = input("Enter product ID he is purchasing from above list: ")

quantity = inventory[prodID]['quantity']
n = inventory[prodID]['name']
price = inventory[prodID]['price']
dis = inventory[prodID]['discount']
profit = inventory[prodID]['profit']
print("Quantity of available goods: ",quantity)
ui_quan = int(input(f"Enter quantity: "))
if ui_quan>quantity:
    print('Invalid Quantity! Try Again.')
else:
    print("\nYour Bill:\n")
    ret = calculate_Bill(price,ui_quan, profit,  dis)
    print(ret[0])
    print('\nAdded to Record. Thanks for purchasing. \nYou can check inventory file now :)')
    inventory[prodID]['quantity'] -= ui_quan
    
    ID = random.randint(105320, 163334)

    if(len(sales)!=0):
        ID = int(list(sales)[-1]) + 1

    sales[ID] = {"Customer":name,"Purchased Item":n,"Quantity":ui_quan,"Total Amount Paid after discount":ret[1],"Profit Earned":ret[2]}
    
    js = json.dumps(inventory)
    with open("inventory.json","w") as fd:
        fd.write(js)
    js = json.dumps(sales)
    with open("sales.json","w") as fd:
        fd.write(js)
```
